### Importing Pandas and DataReader

```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as mp
%matplotlib inline
```

### Checking the Bitcoin price of last year

```python
btc_price= data.DataReader(name='BTC-USD',data_source='yahoo',start='2019-01-01',end= '2019-12-31')
btc_price.head(2)
```

### Plotting the graph only for high and low

```python
btc_price.plot(y = ['High','Low'])
```
